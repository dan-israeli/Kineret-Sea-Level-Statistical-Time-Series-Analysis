The dataset can be found in the project's GitHub repository using the following [link](https://github.com/dan-israeli/Kineret-Sea-Level-Time-Series-Statistical-Analysis/blob/main/Data/Kineret%20Sea%20Level/final_kineret_dataset.csv).

Then, run the code blocks below:

```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
DATASET_PATH = "final_kineret_dataset.csv"
```

```python
dataset = pd.read_csv(DATASET_PATH, index_col="Date")

def display_interval(interval_start, interval_end):
  records = dataset.loc[interval_start:interval_end]

  records.plot(legend=False)
  plt.title(f"""Kineret Level for each Month in Interval
  {interval_start.replace("-", " (") + ")"} - {interval_end.replace("-", " (") + ")"}""")
  plt.xlabel("months")
  plt.ylabel("sea level [meters]")

def get_personalized_graph():
  print("*--- Display Kineret Sea Level In Date Interval ---*\n")
  interval_start = input("Please enter a starting point in format YYYY-MM: ")
  interval_end = input("Please enter an end point in format YYYY-MM: ")
  print()

  display_interval(interval_start, interval_end)
```

Run the code block below in order to launch the user interface. Notice that the input time interval has to be between **1969-01** and **2024-03**.

```python
get_personalized_graph()
```
